### Question 2

```python
# compute the mse and R2

def mse(y, y_pred):
    error = y - y_pred 
    se = np.sum(error**2) 
    mse = se/len(y)  
    return round(mse,3)

def R2(y, y_pred):
    error = y - y_pred 
    se = np.sum(error**2) 
    SSt = np.sum((y - np.mean(y))**2) 
    R2 = 1- (se/SSt) 
    return round(R2,3)
```

```python
ols = LinearRegression().fit(X_train_s, Y_train)
Y_pred = ols.predict(X_test_s)
coef = ols.coef_

print("mse =", mse(Y_test, Y_pred))
print("R2 =", R2(Y_test, Y_pred))
```

### Question 10

```python
n,p = np.shape(scaled_covariates)

selected_variables = []  # initialize empty variables list
remaining_variables = [i for i in range(0,p)]  # initialize result variable list
tests = np.zeros((p,p))
ols = LinearRegression(fit_intercept=False)
residuals = scores

for k in range(p):
    residuals_next = np.zeros((n,p))
    
    for var in remaining_variables:
        
        ols.fit(scaled_covariates.iloc[:,var].values.reshape(-1, 1), residuals)
        residuals_next[:,var] = residuals - ols.predict(scaled_covariates.iloc[:,var].values.reshape(-1, 1))  # update residuals
        
        sigma2 = np.sum(residuals_next[:, var]**2) / (n-1)
        s2 = 1 / np.sum(scaled_covariates.iloc[:, var]**2)
        
        tests[k,var] = np.abs(ols.coef_)/np.sqrt(s2*sigma2)  # compute test values
    
    print("tests", tests)
    
    best_var = np.argmax(tests[k,:])   # choose the next variable

    print("best", best_var)
    remaining_variables.remove(best_var)
    selected_variables.append(best_var)
    residuals = residuals_next[:, best_var]
        
print(selected_variables)
```

### Question 11

```python
n,p = np.shape(scaled_covariates)
selected_variables = []  # initialize empty variables list
selected_p_values = []
remaining_variables = [i for i in range(0,p)]  # initialize result variable list
tests = np.zeros((p,p))
ols = LinearRegression(fit_intercept=False)
p_values = np.ones((p,p))
residuals = scores



for k in range(p):
    residuals_next = np.zeros((n,p))
    
    for var in remaining_variables:
        
        ols.fit(scaled_covariates.iloc[:,var].values.reshape(-1, 1), residuals)
        residuals_next[:,var] = residuals - ols.predict(scaled_covariates.iloc[:,var].values.reshape(-1, 1))  # update residuals
        
        sigma2 = np.sum(residuals_next[:, var]**2) / (n-1)
        s2 = 1 / np.sum(scaled_covariates.iloc[:, var]**2)
        
        tests[k,var] = np.abs(ols.coef_)/np.sqrt(s2*sigma2)  # compute test values
        p_values[k,var] = 2*(1-norm.cdf(tests[k,var]))  # compute p values

    
    print("tests", tests)
    
    best_var = np.argmin(p_values[k,:])   # choose the next variable
    p_value = np.min(p_values[k,:])

    
    print("best", best_var)
    remaining_variables.remove(best_var)
    selected_variables.append(best_var)
    selected_p_values.append(p_value)
    residuals = residuals_next[:, best_var]
        
print(selected_variables, selected_p_values)
```

```python
from scipy.stats import norm

# Add a 1 columns to handle the intercept
x_train_aug = np.column_stack((np.ones((X_train_s.shape[0], 1 )), X_train_s))
p = x_train_aug.shape[1]
n = x_train_aug.shape[0]

resids = Y_train

variables_to_add = list(range(p))
variables_added = []

# Initializes to zeros all the tests/p_values. Costly in memory but more efficient to compute (And needed for what follows)
tests = np.zeros((p,p))
p_values = np.zeros((p,p))

# The intercept is already a 'variable'
reg = LinearRegression(fit_intercept = False)


for k in range(p):
    resids_next = np.zeros((n,p))
    for var in variables_to_add:
        reg.fit(x_train_aug[:,[var]], resids)
        
        # Store the rest for this variable
        resids_next[:,var] = resids - reg.predict(x_train_aug[:,[var]])
        # Compute sigma_2 (1/(n-1) sum(Yi-Ypredi)) times sn1_2 = xTx/n (in R) 
        sigma_x_snk_2 = np.sum(resids_next[:,var]**2)/(n-1)*n/np.sum(x_train_aug[:,var]**2)
        
        # We take the abs of the test directly (as we only need the abs for all the algorithm)
        tests[k, var] = np.sqrt(n) * np.abs(reg.coef_)/np.sqrt(sigma_x_snk_2)
        p_values[k, var] = 2*(1-norm.cdf(tests[k,var]))
    
    best_var = np.argmax(tests[k,:])
    variables_to_add.remove(best_var)
    variables_added.append(best_var)
    resids = resids_next[:,best_var]

print(variables_added)
```
